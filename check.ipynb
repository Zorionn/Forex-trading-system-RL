```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
DF_EUR_USD = pd.read_csv("EUR_USD_2010_2017_bid_ask.csv")
```

```python
DF_EUR_USD['Time'] = pd.to_datetime(DF_EUR_USD['Time'])
```

```python
DF_EUR_USD
```

```python
# DF_EUR_USD = DF_EUR_USD.head(1000)
```

```python
fig, ax1 = plt.subplots(figsize=(10, 6))
ax1.set_xlabel('Time')
ax1.set_ylabel('Bid_Close', color='tab:blue')
ax1.plot(DF_EUR_USD['Time'], DF_EUR_USD['Bid_Close'], color='tab:blue', label='Bid_Close')
ax1.tick_params(axis='y', labelcolor='tab:blue')

ax2 = ax1.twinx()
ax2.set_ylabel('Ask_Close', color='tab:green')
ax2.plot(DF_EUR_USD['Time'], DF_EUR_USD['Ask_Close'], color='tab:green', linestyle='--', label='Ask_Close')
ax2.tick_params(axis='y', labelcolor='tab:green')

plt.title('Historical Data EUR/USD')
fig.tight_layout()  # To prevent overlapping labels
plt.show()
```
